```python
from google.colab import drive
drive.mount('drive', force_remount=True)

densho_narrator = pd.read_csv('/content/drive/My Drive/Colab Notebooks/densho_cleaned_annotated.csv')

densho_narrator = densho_narrator.dropna()

display(densho_narrator)
```

```python
densho_narrator.groupby(["sentiment"])["sentiment"].count()
```

```python
#Multinomial Naive Bayes (MultinomialNB)
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score

X_train, X_test, y_train, y_test = train_test_split(densho_narrator["clean_text"], densho_narrator["sentiment"], test_size=0.8, random_state=0)

vectorizer = CountVectorizer()
model = MultinomialNB()

X_train_counts = vectorizer.fit_transform(X_train)

model.fit(X_train_counts, y_train)

# Make predictions
y_pred = model.predict(vectorizer.transform(X_test))

model_score = cross_val_score(model, X_train_counts, y_train, cv=10)

print("Mean cross-validation score")
print(model_score.mean())

# Evaluate accuracy
print("\n--ACCURACY--\n")
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy score againts the test dataset:\n{accuracy}")

print("\n--RECALL--\n")
recall = recall_score(y_test, y_pred, average="weighted")
print(f"Recall score againts the test dataset:\n{recall}")

print("\n--PRECISION--\n")
precision = precision_score(y_test, y_pred, average="weighted")
print(f"Precision score againts the test dataset:\n{precision}")

print("\n--F-1 SCORE--\n")
f1 = f1_score(y_test, y_pred, average="weighted")
print(f"F-1 score againts the test dataset:\n{f1}")
```

```python
#Random Forest
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score

X_train, X_test, y_train, y_test = train_test_split(densho_narrator["clean_text"], densho_narrator["sentiment"], test_size=0.8, random_state=0)

vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train_count, y_train)

predictions = model.predict(vectorizer.transform(X_test))

model_score = cross_val_score(model, X_train_count, y_train, cv=10)
print("Mean cross-validation score")
print(model_score.mean())

# Evaluate accuracy
print("\n--ACCURACY--\n")
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy score againts the test dataset:\n{accuracy}")

print("\n--RECALL--\n")
recall = recall_score(y_test, predictions, average="weighted")
print(f"Recall score againts the test dataset:\n{recall}")

print("\n--PRECISION--\n")
precision = precision_score(y_test, predictions, average="weighted")
print(f"Precision score againts the test dataset:\n{precision}")

print("\n--F-1 SCORE--\n")
f1 = f1_score(y_test, predictions, average="weighted")
print(f"F-1 score againts the test dataset:\n{f1}")
```

The two algorithms produce consistent result on each metrics. So probably, the problem is in the poorly annotated dataset. For this approach, it would be better if the clustering to be performed preceeding the sentiment analysis.

```python
#cluster the neighborhood
freq = house_train.groupby(["Neighborhood"]).count()["HouseStyle"]
mean = house_train.groupby(["Neighborhood"])["price_per_sqft"].mean()

cluster = pd.concat([freq, mean], axis=1)
cluster['neighborhood'] = cluster.index
cluster.columns = ['freq', 'price_per_sqft','neighborhood']
cluster.describe()

cluster1 = cluster[cluster.price_per_sqft < 101]
print(cluster1.index)

cluster_temp = cluster[cluster.price_per_sqft >= 101]
cluster2 = cluster_temp[cluster_temp.freq <136]
print(cluster2.index)

cluster3 = cluster_temp[cluster_temp.freq >=136]
print(cluster3.index)
```

```python
def get_group(x):
  if x in cluster1.index:
    return 'low_price'
  elif x in cluster2.index:
    return 'high_price_low_freq'
  else:
    return 'high_price_high_freq'

house_train['group'] = house_train["Neighborhood"].apply(get_group)

house_train.head()
```

```python
'''
OverallQual     0.790982
GrLivArea       0.708624
ExterQual       0.682639
KitchenQual     0.659600
GarageArea      0.623431
TotalBsmtSF     0.613581
TotRmsAbvGrd    0.533723
'''
house_train.drop(house_train.columns[[0, 1, 2, 3, 4, 6, 7, 8, 10, 12, 17, 18, 19]], axis=1, inplace=True)
house_train = house_train[['OverallQual', 'GrLivArea', 'ExterQual', 'KitchenQual', 'GarageArea', 'TotalBsmtSF','TotRmsAbvGrd', 'group', 'SalePrice']]
house_train.head()
```

```python
X = house_train[['OverallQual', 'GrLivArea', 'ExterQual', 'KitchenQual', 'GarageArea', 'TotalBsmtSF','TotRmsAbvGrd', 'group']]
Y = house_train[['SalePrice']]

n = pd.get_dummies(house_train.group)
X = pd.concat([X, n], axis=1)

drops = ['group']
X.drop(drops, inplace=True, axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Evaluation
y_pred = regressor.predict(X_test)
print('Linear Regression R squared: %.4f' % regressor.score(X_test, y_test))

import numpy as np
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_pred, y_test)
lin_rmse = np.sqrt(lin_mse)
print('Linear Regression RMSE: %.4f' % lin_rmse)

from sklearn.metrics import mean_absolute_error
lin_mae = mean_absolute_error(y_pred, y_test)
print('Linear Regression MAE: %.4f' % lin_mae)
```

```python
import tensorflow as tf
from transformers import RobertaTokenizer, TFRobertaForSequenceClassification

# Initialize RoBERTa tokenizer
tokenizer = RobertaTokenizer.from_pretrained('roberta-base')

# Sample data (replace this with your own data loading code)
texts = list(X_train)
labels = list(y_train)

# Tokenize texts
tokenized_texts = tokenizer(texts, padding=True, truncation=True, return_tensors="tf")

# Create TensorFlow dataset
dataset = tf.data.Dataset.from_tensor_slices((dict(tokenized_texts), labels))

# Define model
model = TFRobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=3)

# Instantiate optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)

# Compile model
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metrics = ['accuracy']
model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

# Train model
#model.fit(dataset.shuffle(buffer_size=1000).batch(2), epochs=3)
#model.fit(dataset.shuffle(buffer_size=1000).batch(4), epochs=4)
model.fit(dataset.shuffle(buffer_size=1000).batch(2), epochs=3)
```
