```python
from google.colab import drive
drive.mount('drive', force_remount=True)

densho_narrator = pd.read_csv('/content/drive/My Drive/Colab Notebooks/densho_cleaned_annotated.csv')

densho_narrator = densho_narrator.dropna()

display(densho_narrator)
```

```python
densho_narrator.groupby(["sentiment"])["sentiment"].count()
```

```python
#Multinomial Naive Bayes (MultinomialNB)
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score

X_train, X_test, y_train, y_test = train_test_split(densho_narrator["clean_text"], densho_narrator["sentiment"], test_size=0.8, random_state=0)

vectorizer = CountVectorizer()
model = MultinomialNB()

X_train_counts = vectorizer.fit_transform(X_train)

model.fit(X_train_counts, y_train)

# Make predictions
y_pred = model.predict(vectorizer.transform(X_test))

model_score = cross_val_score(model, X_train_counts, y_train, cv=10)

print("Mean cross-validation score")
print(model_score.mean())

# Evaluate accuracy
print("\n--ACCURACY--\n")
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy score againts the test dataset:\n{accuracy}")

print("\n--RECALL--\n")
recall = recall_score(y_test, y_pred, average="weighted")
print(f"Recall score againts the test dataset:\n{recall}")

print("\n--PRECISION--\n")
precision = precision_score(y_test, y_pred, average="weighted")
print(f"Precision score againts the test dataset:\n{precision}")

print("\n--F-1 SCORE--\n")
f1 = f1_score(y_test, y_pred, average="weighted")
print(f"F-1 score againts the test dataset:\n{f1}")
```

```python
#Random Forest
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score

X_train, X_test, y_train, y_test = train_test_split(densho_narrator["clean_text"], densho_narrator["sentiment"], test_size=0.8, random_state=0)

vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train_count, y_train)

predictions = model.predict(vectorizer.transform(X_test))

model_score = cross_val_score(model, X_train_count, y_train, cv=10)
print("Mean cross-validation score")
print(model_score.mean())

# Evaluate accuracy
print("\n--ACCURACY--\n")
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy score againts the test dataset:\n{accuracy}")

print("\n--RECALL--\n")
recall = recall_score(y_test, predictions, average="weighted")
print(f"Recall score againts the test dataset:\n{recall}")

print("\n--PRECISION--\n")
precision = precision_score(y_test, predictions, average="weighted")
print(f"Precision score againts the test dataset:\n{precision}")

print("\n--F-1 SCORE--\n")
f1 = f1_score(y_test, predictions, average="weighted")
print(f"F-1 score againts the test dataset:\n{f1}")
```

The two algorithms produce consistent result on each metrics. So probably, the problem is in the poorly annotated dataset. For this approach, it would be better if the clustering to be performed preceeding the sentiment analysis.

```python
# Write your code here
import pandas as pd
from google.colab import drive
drive.mount('drive', force_remount=True)

densho_narrator = pd.read_csv('/content/drive/My Drive/Colab Notebooks/densho_cleaned_annotated.csv')

densho_narrator = densho_narrator.dropna()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, x_test = train_test_split(densho_narrator["clean_text"], densho_narrator["sentiment"], test_size = 0.8, random_state = 0)

X_train = X_train.tolist()
X_test = X_test.tolist()
y_train = y_train.tolist()
y_test = x_test.tolist()
```

```python
!pip install transformers
!pip install accelerate -U
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
import numpy as np
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments
import accelerate
import transformers

training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)

model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset             # evaluation dataset
)

#trainer.train()
```
